```python
from langgraph.graph import StateGraph, START , END
from typing import TypedDict
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
model= ChatGroq(model="llama-3.1-8b-instant")
```

```python
model= ChatGroq(model="llama-3.1-8b-instant")
```

```python
class BlogState(TypedDict):
    title: str
    outline: str
    content: str
```

```python
def create_outline(state: BlogState) -> BlogState:

    # fetch title
    title = state['title']

    # call llm gen outline
    prompt = f"Create a detailed outline for a blog post about {title}"
    outline= model.invoke(prompt).content
    # update state
    state['outline'] = outline

    return state
```

```python
def create_blog(state: BlogState) -> BlogState:

    # fetch title and outline
    title = state['title']
    outline = state['outline']

    # call llm gen blog content
    prompt = f"Write a blog post about {title} based on the following outline: {outline}"
    content= model.invoke(prompt).content
    # update state
    state['content'] = content

    return state
```

```python
graph= StateGraph(BlogState)

#nodes
graph.add_node('create_outline',create_outline)
graph.add_node('create_blog',create_blog)

# Edges
graph.add_edge(START, 'create_outline')
graph.add_edge('create_outline', 'create_blog')    
graph.add_edge('create_blog', END)

workflow = graph.compile()
```

```python
# Corrected: use string key and proper variable name
initial_state = {"title": "Rise of Palantier"}
final_state = workflow.invoke(initial_state)
print(final_state)
```
